```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Matrisin ölçüsünü istifadəçidən al
n = int(input("Matrisin ölçüsünü daxil edin (n x n): "))
print("Matrisin elementlərini sətir-sətir daxil edin:")
matrix = []
for i in range(n):
    row = list(map(float, input(f"{i+1}. sətir: ").split()))
    matrix.append(row)

# Matrisi NumPy array-ə çevir
matrix = np.array(matrix)

# eigenvalues, eigenvectors hesablanır
eigenvalues, eigenvectors = np.linalg.eig(matrix)

# Nəticələri ekrana çıxart
print("EigenValue:")
print(eigenvalues)

print("\nEigenVector:")
print(eigenvectors)

# 3D qrafik yarat
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# EigenVector 3D məkanında göstər
for i in range(n):
    # EignVector miqyasını tənzimləyirik
    ax.quiver(0, 0, 0, eigenvectors[0, i], eigenvectors[1, i], eigenvectors[2, i],
              length=1, color='blue', arrow_length_ratio=0.1)

# Oxların limitlərini ayarlamaq
ax.set_xlim([-2, 2])
ax.set_ylim([-2, 2])
ax.set_zlim([-2, 2])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Qrafikin başlığını və grid-i tənzimləmək
ax.set_title('EigneVector 3D (Miqyaslı)')
ax.grid(True)

# Qrafiki göstərmək
plt.show()
```

```python
import sympy as sp

def get_eigenvalues_and_eigenvectors(matrix):
    A = sp.Matrix(matrix)

    # Özqiymətləri tapırıq
    eigenvalues = A.eigenvals()

    # Özvektorları tapırıq
    eigenvectors = A.eigenvects()

    # Normalizasiya olunmuş özvektorları qaytarmaq üçün siyahı yaradırıq
    normalized_eigenvectors = []

    for eigval, multiplicity, eigvecs in eigenvectors:
        for vec in eigvecs:
            norm_vec = vec / vec.norm()
            normalized_eigenvectors.append(norm_vec)

    return list(eigenvalues.keys()), normalized_eigenvectors

# İstifadəçidən 3x3 matrisin elementlərini daxil etməyi istəyirik
matrix = []
print("3x3 ölçülü matrisin elementlərini sətir-sətir daxil edin:")

for i in range(3):
    row = list(map(float, input(f"{i+1}-ci sətri daxil edin (3 ədəd, boşluqla ayırın): ").split()))
    if len(row) != 3:
        print("Xəta! Zəhmət olmasa, 3 ədəd daxil edin.")
        exit()
    matrix.append(row)

# Özqiymətləri və özvektorları hesablayırıq
eigenvalues, eigenvectors = get_eigenvalues_and_eigenvectors(matrix)

# Nəticələri çap edirik
print("\nMatrisin özqiymətləri (eigenvalues):")
print(eigenvalues)

print("\nMatrisin normallaşdırılmış özvektorları (eigenvectors):")
for vec in eigenvectors:
    print(vec)
```
